```python
#prepare the data 
import pandas as pd
data = pd.read_excel("features_cleaned.xlsx", engine="openpyxl")
x=data[['IsHoliday','Temperature','Fuel_Price','CPI','Unemployment']]
y=data['Weekly_Sales']
```

```python
#split the data 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.7,random_state=42)
```

```python
#Train a linear regression model
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
import numpy as np
print(np.isnan(x_train).sum())
print(np.isnan(y_train).sum())
imputer = SimpleImputer(strategy='mean')  # You can also use 'median'
x_train = imputer.fit_transform(x_train)
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
#predict on test data

import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression

# Convert to DataFrame if not already
x_train = pd.DataFrame(x_train)
x_test = pd.DataFrame(x_test)

# Impute missing values
imputer = SimpleImputer(strategy='mean')
x_train = imputer.fit_transform(x_train)
x_test = imputer.transform(x_test)

# Fit the model
model = LinearRegression()
model.fit(x_train, y_train)

# Predict using the model
y_pred = model.predict(x_test)


```

```python
#evaluate the model

mae=mean_absolute_error(y_test,y_pred)
rmse=np.sqrt(mean_squared_error(y_test,y_pred))
print(f"MAE:{mae},RMSE:{rmse}")             
```

```python
#insights and visualization

#feature importance visualization (for linear regression)
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Convert to DataFrame if not already
x_train_df = pd.DataFrame(x_train)
x_test_df = pd.DataFrame(x_test)

# Store the column names
columns = x_train_df.columns

# Impute missing values
imputer = SimpleImputer(strategy='mean')
x_train = imputer.fit_transform(x_train_df)
x_test = imputer.transform(x_test_df)

# Convert back to DataFrame to retain column names
x_train_df = pd.DataFrame(x_train, columns=columns)
x_test_df = pd.DataFrame(x_test, columns=columns)

# Fit the model
model = LinearRegression()
model.fit(x_train_df, y_train)

# Predict using the model
y_pred = model.predict(x_test_df)

# Define features using the original DataFrame
features = columns

# Feature importance visualization
importance = model.coef_
plt.bar(features, importance)
plt.title("Feature Importance")
plt.xlabel("Features")
plt.ylabel("Importance")
plt.xticks(rotation=90)
plt.show()

```
